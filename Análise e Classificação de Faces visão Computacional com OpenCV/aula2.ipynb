```python
import cv2 as cv
import matplotlib.pyplot as plt

```

```python
imagem = cv.imread("imagens/px-people.jpg")
imagem = cv.cvtColor(imagem, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(imagem)
```

```python
imagem_gray = cv.cvtColor(imagem, cv.COLOR_RGB2GRAY)
```

```python
classificador = cv.CascadeClassifier(
    'classificadores\haarcascade_frontalface_default.xml')
```

```python
faces = classificador.detectMultiScale(imagem_gray, 1.3, 5)
```

```python
len(faces)
```

```python
faces[0]
```

```python
imagem_anotada = imagem.copy()
```

```python
for (x,y,w,h) in faces:
    cv.rectangle(imagem_anotada, (x,y), (x+w, y+h), (255, 255, 0),2)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(imagem_anotada)
```

```python
face_imagem = 0

for (x,y,w,h) in faces:
    face_imagem += 1
    image_roi = imagem[y:y+h, x:x+w]
    image_roi = cv.cvtColor(image_roi, cv.COLOR_RGB2BGR)
    cv.imwrite("face_{}.png".format(face_imagem), image_roi)
```
